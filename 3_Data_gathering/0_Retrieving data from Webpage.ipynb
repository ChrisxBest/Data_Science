# **1. Retrieving data from Webpage**

```python
import requests
website_url = 'http://web.archive.org/web/20240401160228/https://en.wikipedia.org/wiki/DAX'
```

```python
response = requests.get(website_url)
print(response.status_code == requests.codes.ok)
```

```python
response.text[:1530]
```

```python
response.text.count('<table')
```

# **2. Reading HTML-Doc**

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text)
```

```python
print(soup.prettify()[:1575])
```

```python
tables = soup.find_all('table')
len(tables)
```

```python
for table in tables:
    print(table.attrs)
```

```python
table = soup.find(id='constituents')
```

```python
table.text
```

```python
table_list = []
for row in table.find_all('tr'):
    table_list.append(row.text.split('\n'))
print(table_list)
```

```python
for row in table_list:
    print(len(row))
```

```python
import pandas as pd
df_dax = pd.DataFrame(table_list[1:], columns=table_list[0])
print(df_dax.shape)
df_dax.head()
```

```python
df_dax = df_dax.drop('', axis=1)
df_dax.head()
```

```python
df_dax.info()
```

```python
df_dax.loc[:, 'Employees'] = df_dax.loc[:, 'Employees'].str.replace(r'\(\d\d\d\d\)', '', regex=True)
```

```python
df_dax.loc[:, 'Index weighting (%)1'] = pd.to_numeric(df_dax.loc[:, 'Index weighting (%)1'], errors='coerce')
df_dax.loc[:, 'Employees'] = pd.to_numeric(df_dax.loc[:, 'Employees'].str.replace(',', ''))
df_dax.loc[:, 'Founded'] = pd.to_numeric(df_dax.loc[:, 'Founded'])
```
